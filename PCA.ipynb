```python
from sklearn.decomposition import PCA
```

```python
from sklearn.datasets import load_iris as data
```

```python
irisdata=data().data
```

```python
irisdata
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x=StandardScaler().fit_transform(irisdata)
```

### scaling done

```python
# convert data set in pca
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(x)
```

```python
df.head()
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit_transform(x)
```

```python
pcadf=pd.DataFrame(pca.fit_transform(x))
```

```python
pcadf
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df[[0]],df[[1]])
```

```python
plt.scatter(pcadf[[0]],pcadf[[1]])
```

```python
#doing learning on without pca
```

```python
from sklearn import neighbors
```

```python
clf=neighbors.KNeighborsClassifier(n_neighbors=5)
```

```python
clf.fit(df,data().target)
```

```python
clf.score(df,data().target)
```

```python
# now using pca
```

```python
model=neighbors.KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(pcadf,data().target)
```

```python
model.score(pcadf,data().target)
```
